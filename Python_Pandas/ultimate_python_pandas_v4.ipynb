```python
import pandas as pd 
athlest_data = pd.read_csv('DataSource/Athlets_information_modified.csv')
```

```python
athlest_data
```

```python
athlest_data_new = athlest_data.copy()
```

```python
athlest_data_new
```

```python
# create a category of "Height_category" based on certain conditions of athlets height, as Average,Tall and Short
athlest_data_new['Height_category'] = athlest_data_new['Height_cm'].apply(lambda x: 'Short' if x<=170 else ('Average' if x <=175 else 'Tall'))
```

```python
athlest_data_new.head()
```

```python
# define function to make categorize of athlets based on their heights and weights 
def categorize_athlet(row):
    if row['Height_cm'] <= 170 and row['Weight_kg'] <=70: 
        return 'Light-weight' 
    elif row['Height_cm'] <175 and row['Weight_kg'] <=78:
        return 'Middle-weight' 
    else: 
        return 'Heavy-weight' 
athlest_data_new['Category'] = athlest_data_new.apply(categorize_athlet, axis=1)
```

```python
athlest_data_new
```

```python
athlest_data_new['Athlets_id'] = athlest_data_new['Athlete_id']+1000
```

```python
athlest_data_new
```

```python
athlest_data_new.drop(columns='Athlete_id', inplace=True)
```

```python
athlest_data_new
```

```python
athlest_data_new.rename(columns={'Athlets_id': 'Athletes_id'})
```

```python
coffee_data = pd.read_csv('DataSource/modified_coffee.csv')
```

```python
coffee_data.drop(columns='Unnamed: 0', inplace=True)
```

```python
coffee_data
```

```python
coffee_data['Price'] = 5.99
```

```python
coffee_data
```

```python
coffee_data['Revenue'] = coffee_data['Price'] * coffee_data['Units Sold']
```

```python
coffee_data
```

```python
# find out total revenue earned by the each type of 'Coffee Type' sold. 
coffee_data.groupby('Coffee Type')['Revenue'].sum()
```

```python
# find out total revenue earned by the each type of 'Coffee Type' sold. 
coffee_data.groupby('Coffee Type')['Revenue'].sum().reset_index()
```

```python
# find out total revenue earned based on the each weekdays 
coffee_data.groupby('Day')['Revenue'].sum().reset_index()
```

```python
# find out total revenue earned by the each type of 'Coffee Type' sold and what number of each type of coffee sold. 
coffee_data.groupby('Coffee Type')[['Revenue','Units Sold']].sum().reset_index()
```

```python
# find out total what number of units sold for each type of coffee type.  
coffee_data.groupby('Coffee Type')['Units Sold'].sum().reset_index()
```

```python
# i want to know which type of coffee sold in which day in what number of units sold
coffee_data.groupby(['Coffee Type','Day'])['Units Sold',].sum().reset_index()
```

```python
# find out in which day which type of coffee sold most
coffee_data.groupby(['Day','Coffee Type'])['Units Sold',].sum().reset_index()
```

```python
coffee_data.groupby(['Day'])['Revenue'].sum().reset_index()
```

```python
coffee_data.to_csv('DataSource/modified_coffee.csv', index_label=False, index=False)
```
