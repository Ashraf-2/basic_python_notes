```python
sales_data_copy['First_Name'] = sales_data_copy['full_names'].str.split(' ').str[0]
```

```python
sales_data_copy['Transection_date_object'] = pd.to_datetime(sales_data_copy['Transection_date'])
```

```python
sales_data_copy
```

```python
sales_data_copy['Transection_Year'] = sales_data_copy['Transection_date_object'].dt.year
sales_data_copy
```

```python
sales_data_copy.drop(columns="Transection_date_object", inplace=True)
```

```python
sales_data_copy
```

```python
# filter data based on products, regions and transection year 
sales_data_copy[['products','regions','Transection_Year']]
```

```python
# find out only the regions of where leptops has been sold and the revenue 
sales_data_copy.loc[sales_data_copy['products'] == 'Laptop']
```

```python
# find out only the regions of where tablet has been sold and earned reveneu less then 950 and the revenue 
sales_data_copy.loc[(sales_data_copy['products'] == 'Tablet') &  (sales_data_copy['Revenue'] < 950)]
```

```python
# Task: find out the sum of reveneu in every unique regions and show it using %matplotlib .
group_region_revenue = sales_data_copy.groupby('regions')['Revenue'].sum().reset_index()
group_region_revenue
```

```python
regions_list = list(group_region_revenue['regions'])
regions_list
```

```python
revenueu_list = list(group_region_revenue['Revenue'])
revenueu_list
```

```python
import matplotlib.pyplot as plt 
# plt.plot(['Asia', 'Australia', 'Europe', 'North America', 'South America'], [4300, 4660, 4210, 3960, 2820])
plt.plot(regions_list,revenueu_list)
plt.xlabel("Sales Regions")
plt.ylabel("Total Revenue") 
plt.title("Total Sales in Different Regions") 
plt.show()
```
