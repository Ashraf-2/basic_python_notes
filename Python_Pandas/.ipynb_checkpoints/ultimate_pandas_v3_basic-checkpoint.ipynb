```python
import pandas as pd 
```

```python
coffee_data = pd.read_csv('DataSource/coffee.csv')
```

```python
coffee_data
```

```python
# Add one column in the existing dataFrame, example add "price" column having 5 in values. 
coffee_data['Price'] = 7
```

```python
coffee_data
```

```python
# using numpy add the 'new price' column and if the 'Coffee type' is "Espresso" then add the value 4.9 otherwise 5.9 in this "new price" column 
import numpy as np 
coffee_data['New_Price'] = np.where(coffee_data['Coffee Type'] == 'Espresso',4.99, 5.99)
```

```python
coffee_data
```

```python
# drop (delete) the 0th index of raw. 
coffee_data.drop(0) 
# but it will not modified our main dataframe , it just return us to show as according to our code
```

```python
# drop "Price" column from the dataframe 
coffee_data.drop(columns="Price")
```

```python
coffee_data['Price'] = 7.50 
coffee_data
```

```python
# truely drop the "Price" column from the dataframe, it should be deleted. not just to return. 
coffee_data.drop(columns="Price", inplace=True) # inplace -> true means you will delete the column permanently
```

```python
athlets_data_new.head()
```

```python

athlets_data_new.rename(columns={'First Name': "First_Name"}, inplace=True)
```

```python
# Task: find out if there is any athlets having name "Ali". 
athlets_data_new.query('First_Name == "Ali"')
```

```python
athlets_data_new.info()
```

```python
# From "Born Date" find out each of athlets "Born_Year" 
# 1. at first check the data type of "born date" field. 
# 2. now convert it into date type data. 
# 3. now split the year from the date. 

athlets_data_new['Born_Date_Time'] = pd.to_datetime(athlets_data_new['Born_Date'], format="%Y-%m-%d") # creating new dataTime field 

athlets_data_new['Born_Year'] = athlets_data_new['Born_Date_Time'].dt.year # creating a filed "Born_Year" from the created datetime data type of column.

athlets_data_new
```
