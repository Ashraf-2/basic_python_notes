```python
import pandas as pd
```

```python
import requests
import json
```

```python
data = requests.get("https://fruityvice.com/api/fruit/all")
```

```python
results = json.loads(data.text)
```

```python
fruits_nutritions = pd.DataFrame(results)
```

```python
fruits_nutritions
```

```python
fruits_nutritions.to_csv('DataSource/Fruit_nutritions.csv', index_label=False, index=False)
```
